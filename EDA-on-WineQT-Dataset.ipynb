```python
import seaborn as sns
import matplotlib.pyplot as plt


```

#Vis - 1
#telling quality based on rating

```python
sns.countplot(x='quality', data=df, palette='coolwarm')
plt.title('Distribution of Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Count')
plt.show()
```

#VIS - 2
#Box Plot for Alcohol Content across Quality Ratings

```python
sns.boxplot(x='quality', y='alcohol', data=df, palette='Set2')
plt.title('Alcohol Content across Quality Ratings')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show()

```

#VIS 3

#Scatter Plot for Alcohol Content vs. Wine Quality

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['alcohol'], df['quality'], alpha=0.6, c=df['quality'], cmap='viridis')
plt.colorbar(label='Quality')
plt.title('Alcohol Content vs. Wine Quality')
plt.xlabel('Alcohol Content')
plt.ylabel('Quality')
plt.grid(True)
plt.show()
```

#VIS - 4

#Correlation Heatmap of Features

```python
corr = df.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Features')
plt.show()

```
